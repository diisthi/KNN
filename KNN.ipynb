```python
import pandas as pd
import numpy as np
```

## Data Prosessing

```python
data = pd.read_csv("diabetes-dataset.csv")
data
```

```python
data.info()
```

```python
data.corr()
```

```python
x = data.drop("Outcome", axis = 1)
y = data["Outcome"]
```

```python
x
```

## Split Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size = 0.25,
                                                    random_state=0)
```

## Model KKN

```python
from sklearn.neighbors import KNeighborsClassifier
knnModel=KNeighborsClassifier(n_neighbors = 5)
knnModel.fit(x_train,y_train)
```

```python
predict = knnModel.predict(x_test)
predict
```

## Model Evaluation

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print(classification_report(y_test, predict))
```

```python
print("K-Nearst Neighbors model accuracy (in %) :", accuracy_score(y_test,predict)*100)
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

cm = metrics.confusion_matrix(y_test, predict)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=knnModel.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsTransformer
import numpy as np

X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

# Membuat instance KNeighborsTransformer
n_neighbors = 2
knn_transformer = KNeighborsTransformer(n_neighbors=n_neighbors, mode='distance')

# Mengubah data menggunakan KNeighborsTransformer|
X_transformed = knn_transformer. fit_transform(X)

print("Data Asli:")
print(X)

print("\nData yang Telah Diubah:")
print(X_transformed)
```
